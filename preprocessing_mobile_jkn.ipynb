## Preprocessing

case folding

```python
def case_folding(text):
  if isinstance(text, str):
    lowercase_text = text.lower()
    return lowercase_text
  else:
    return text

df['case_folfing'] = df['content'].apply(case_folding)
df.head(10)
```

cleaning url, html, emoji, number, and symbols

```python
import re
import string
import nltk

def remove_url(youtube):
  url = re.compile(r'https?://\S+|www\.\S+')
  return url.sub(r'', youtube)

def remove_html(youtube):
  html = re.compile(r'<.*?>')
  return html.sub(r'', youtube)

def remove_emoji(youtube):
  emoji_pattern = re.compile("["
  u"\U0001F600-\U0001F64F"
  u"\U0001F300-\U0001F5FF"
  u"\U0001F680-\U0001F6FF"
  u"\U0001F1E0-\U0001F1FF"
  u"\U00002500-\U00002BEF"
  u"\U00002702-\U000027B0"
  u"\U00002702-\U000027B0"
  u"\U000024C2-\U0001F251"
  u"\U0001f926-\U0001f937"
  u"\U00010000-\U0010ffff"
  u"\u2640-\u2642"
  u"\u2600-\u2B55"
  u"\u200d"
  u"\u23cf"
  u"\u23e9"
  u"\u231a"
  u"\ufe0f"
  u"\u3030"
  "]+", flags=re.UNICODE)
  return emoji_pattern.sub(r'', youtube)

def remove_number(youtube):
  return re.sub(r'\d+', '', youtube)

def remove_symbol(youtube):
  return youtube.translate(str.maketrans('', '', '!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~‚Äú‚Äù"'))

df['clean'] = df['case_folfing'].apply(lambda x: remove_url(x))
df['clean'] = df['clean'].apply(lambda x: remove_html(x))
df['clean'] = df['clean'].apply(lambda x: remove_emoji(x))
df['clean'] = df['clean'].apply(lambda x: remove_number(x))
df['clean'] = df['clean'].apply(lambda x: remove_symbol(x))

df.head(15)
```

tokenization

```python
def tokenization(text):
  tokens = text.split()
  return tokens

df['tokenisasi'] = df['clean'].apply(tokenization)
df.head(10)
```

stopword removal

```python
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
factory = StopWordRemoverFactory()
stop_words = factory.get_stop_words()
print(stop_words)
```

```python
def remove_stopwords(text):
  return [word for word in text if word not in stop_words]

df['stopwords_remove'] = df['tokenisasi'].apply(lambda x: remove_stopwords(x))
df.head(10)
```

stemming

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
```

```python
stemmed = StemmerFactory()
stemmer = stemmed.create_stemmer()

def text_stem(text):
  return [stemmer.stem(word) for word in text]

df['stemmed'] = df['stopwords_remove'].apply(lambda x: ' '.join(text_stem(x)))
print(df.head(10))
```

```python
df.head(10)
```

split text

```python
def split_text(text):
  tokens = text.split()
  return tokens

df['preprocessed'] = df['stemmed'].apply(split_text)
df.head(10)
```
